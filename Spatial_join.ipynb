## Attach gauge IDs to Polygons for latest spatial join product
I had to manually do a spatial join in ArcPro, attaching the IDs to the names of the reservoir. This left me with a csv file of names alongside corresponding gauge IDs, but the data wasn't spatial. Here I add the spatial element (polygons with names) to the gauge IDs. The common variable to merge on is the reservoir names. 

```python
import pandas as pd
import geopandas as gpd
```

```python
gdf = gpd.read_file('Named_Reservoirs/Named_Reservoirs4_with_ID.shp')
df = pd.read_csv('guage_to_reservoir_name_join.csv')
gdf2 = gdf.merge(df, on = 'NAME')
gdf2
```

```python
gdf2.to_file('Named_Reservoirs4_with_ID.shp')
```
